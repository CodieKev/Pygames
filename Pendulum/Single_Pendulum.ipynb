```python

pygame.init()

xmin = ymin = 0
xmax = ymax = 500

win = pygame.display.set_mode((xmax,ymax))
pygame.display.set_caption("Pendelum")

theta = -pi/3
w = 0
l = 500/3
x = 250+l*sin(theta)
y = (500/3)*(1+cos(theta))
radius = thickness = 20
velx = 15
vely=0
elastic_coeff = 0.85
g =1# 980*ratio
t = 10
run = True
time = 1

dt = 0.001
def f1(x,v,t):
	dxdt =v
	return dxdt
def f2(x,v,t):
	dvdt = -g*sin(theta)
	return dvdt
def rk4(x,v,t):
	k11 = dt*f1(x,v,t);
	k21 = dt*f2(x,v,t);
	k12 = dt*f1(x+0.5*k11,v+0.5*k21,t+0.5*dt);
	k22 = dt*f2(x+0.5*k11,v+0.5*k21,t+0.5*dt);
	k13 = dt*f1(x+0.5*k12,v+0.5*k22,t+0.5*dt);
	k23 = dt*f2(x+0.5*k12,v+0.5*k22,t+0.5*dt);
	k14 = dt*f1(x+k13,v+k23,t+dt);
	k24 = dt*f2(x+k13,v+k23,t+dt);
	delx = (k11+2*k12+2*k13+k14)/6;
	delv = (k21+2*k22+2*k23+k24)/6;
	return delx 
def rk41(x,v,t):
	k11 = dt*f1(x,v,t);
	k21 = dt*f2(x,v,t);
	k12 = dt*f1(x+0.5*k11,v+0.5*k21,t+0.5*dt);
	k22 = dt*f2(x+0.5*k11,v+0.5*k21,t+0.5*dt);
	k13 = dt*f1(x+0.5*k12,v+0.5*k22,t+0.5*dt);
	k23 = dt*f2(x+0.5*k12,v+0.5*k22,t+0.5*dt);
	k14 = dt*f1(x+k13,v+k23,t+dt);
	k24 = dt*f2(x+k13,v+k23,t+dt);
	delx = (k11+2*k12+2*k13+k14)/6.0;
	delv = (k21+2*k22+2*k23+k24)/6.0;
	return delv 



x1,y1 = pygame.mouse.get_pos()

while run:
    pygame.time.delay(time)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
    
    
    
    
    
    if pygame.mouse.get_pressed()[0] and abs(x1-x)<25 and abs(y1-y)<25:
        #x1,y1 =pygame.mouse.get_pos()
        x,y = pygame.mouse.get_pos()
        w = 0
        l = ((x-250)**2+(y-500/3)**2)**0.5
        theta1 = asin((x-250)/l)
        theta2 = acos((y-500/3)/l)
        if theta1 == theta2:
            theta = theta2
        elif 0<theta1<theta2:
            theta = theta2
        elif -pi/2<theta1<0 and 0<theta2<pi/2:
            theta = theta1
        else:
            theta = -pi/2-abs(-pi/2-theta1)
        x1,y1 = pygame.mouse.get_pos()
    else:
        theta = theta + rk4(theta,w,t)
        w = w + rk41(theta,w,t)
        #w = 0.9999*w + rk41(theta,w,t)      # Damped

        x = 250+l*sin(theta)
        y = (500/3+l*cos(theta))
        x1,y1 = pygame.mouse.get_pos()
    win.fill((255,255,255))
     
    pygame.draw.line(win, (105,105,105), (250,500/3), (x,y), width=3)
    pygame.draw.circle(win, (0,0,0), (250, 500/3), 5, 5)
    pygame.draw.circle(win, (0,0,0), (x, y), radius, radius)
    pygame.display.update() 
    
pygame.quit()
```
